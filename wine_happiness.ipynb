```python
# import the dependencies
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
import pymysql
pymysql.install_as_MySQLdb()
```

```python
# import the new excel file and add to a database
excel_file = "data/happiness.xls"
happiness_df = pd.read_excel(excel_file)
happiness_df.head()
```

```python
# verifty the length of the columns (can not see all of them)
len(happiness_df.columns)
```

```python
# Create a new database with only the data we want which helps create the happiness index 
new_happiness_df = happiness_df[[
    'Country name', 
    'Year', 
    'Log GDP per capita', 
    'Social support',
    'Healthy life expectancy at birth',
    'Freedom to make life choices',
    'Generosity',
    'Perceptions of corruption'
]]
```

```python
# check the database
new_happiness_df.head()
```

```python
# rename the columns of the database
new_happiness_df = new_happiness_df.rename(columns={
    "Country name": "country_name", 
    "Year": "year",
    "Log GDP per capita": "log_gdp_per_capita",
    "Social support": "social_support",
    "Healthy life expectancy at birth": "healthy_life_expectancy_at_birth",
    "Freedom to make life choices": "freedom_to_make_life_choices",
    "Generosity": "generosity",
    "Perceptions of corruption": "perceptions_of_corruption"
    }
)

new_happiness_df.head()
```

```python
# import the 2nd excel file and add to a database
excel_file2 = "data/wine_production_by_country.xlsx"
wine_production_df = pd.read_excel(excel_file2)
wine_production_df.head()
```

```python
# rename the columns
wine_production_df = wine_production_df.rename(columns={
    "Country": "country_name", 
    "Year": "year",
    "Wine production in mhl": "wine_production_in_mhl",
    }
)
wine_production_df.head()
```

```python
# create a connection instance to local MySQL instance
rds_connection_string = "root:root@127.0.0.1/etl_project_db"
engine = create_engine(f'mysql://{rds_connection_string}')
```

```python
# verify the table names
engine.table_names()
```

```python
# add the Pandas dataframe for wine production into the our SQL table
wine_production_df.to_sql(name='wine_production', con=engine, if_exists='append', index=False)
```

```python
# add the Pandas dataframe for happiness into the our SQL table
new_happiness_df.to_sql(name='happiness', con=engine, if_exists='append', index=False)
```
